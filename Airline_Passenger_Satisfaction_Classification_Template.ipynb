````python
# Numerical Columns

num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical Columns
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

numeric_columns = df.select_dtypes(include=np.number).columns
for col in numeric_columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
````

Outlier Treatment using IQR

````python
for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower,
                       lower,
                       df[col])
    df[col] = np.where(df[col] > upper,
                       upper,
                       df[col])
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

*   The Interquartile Range (IQR) method was used to identify outliers.
*   Extreme values were capped using the lower and upper IQR limits instead of deleting observations.
*   This approach preserves valuable information while reducing the impact of extreme values on machine learning models.
